```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Step 1: Load data from Excel file
file_path = '/content/svm (1).xlsx'
data = pd.read_excel(file_path)
X = data['X'].values
y = data['y'].values

# Step 2: Initialize model parameters
num_iterations = 1000
learning_rate = 0.01
m = len(y)
w = np.zeros(2)  # Initialize weights [w0, w1] as 0

# Step 3: Implement gradient descent
for iteration in range(num_iterations):
    # Calculate predictions (hypothesis)
    y_pred = w[0] + w[1] * X

    # Calculate the cost (J(w))
    cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)

    # Calculate the gradient for w0 and w1
    gradient_w0 = (1 / m) * np.sum(y_pred - y)
    gradient_w1 = (1 / m) * np.sum((y_pred - y) * X)

    # Update weights using the gradient
    w[0] = w[0] - learning_rate * gradient_w0
    w[1] = w[1] - learning_rate * gradient_w1

    # Print cost for every 100 iterations
    if iteration % 100 == 0:
        print(f'Iteration {iteration}: J(w) = {cost}')

# Step 4: Print the final weights and cost
print(f'Final weights: w0 = {w[0]}, w1 = {w[1]}')
print(f'Final cost: J(w) = {cost}')

# Step 5: Plot the data and regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data Points', color='blue')
plt.plot(X, w[0] + w[1] * X, label='Regression Line', color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.grid(True)
```
