```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,1,10)
m=0.8
c=30
y=m*x+c
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=6+3*np.sin(2*x)

fig,a=plt.subplots()
a.plot(x,y,linewidth=2.0)

a.set(xlim=(0,8), xticks=np.arange(1,8),
      ylim=(0,8), yticks=np.arange(1,8))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,600])

plt.title("Thermistor characteristics")
plt.xlabel("Temperature")
plt.ylabel("Resistance")

plt.plot(temp,res)
plt.grid()
plt.show()
```
