```python
def cake_baker(p,f,s):
  print("No of pieces:",p)
  print("Flavour:",f)
  print("order size:",s)
cake_baker("5","Chocolate","3kg")
cake_baker("25","Vanilla","5kg")
```

```python
def add(*args):
  sum=0
  for i in args:
    sum=sum+i
  return sum
print(add(1,2,9))
```

```python
def greet(**kwargs):
  print("hello")
  for key, value in kwargs.items():
    print(value, end=" ")

greet(name="pranav",usn="01fe21bec025",branch="ECE")
```

```python
import random
computer=random.randint(1,9)
userinput=int(input("Enter the number between 1 and 9"))
print(computer)
print(userinput)

if computer>userinput:
  print("Computer won")
elif computer==userinput:
  print("TIE")
else:
  print("you won")
```

```python
import random
gamelist=['rock','paper','scissor']
z=random.choice(gamelist)
user=input("Enter the choice")
print(z)
print(user)

if z==user:
  print("TIE")
elif (z=="rock" and user=="paper"):
  print("computer won")
elif(z=="paper" and user=="scissor"):
  print("user won")
elif(z=="scissor" and user=="paper"):
  print("computer won")
```
