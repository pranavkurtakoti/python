Looping statements

while and for loop

```python
num=int(input("Enter the number:"))
while num<=10:
  print("Hello world")
  break
while num>10 and num<=99:
  print("Bye")
  break
```

```python
num=int(input("Enter the number"))
for num in range(num):
  print("single digit number")
for num in range(num,12,3):
  print("numbers")
```

```python
places=["goa","havelock","mauritius"]
for x in places:
  print(x)
  if x=="mauritius":
    print("found")
    break
```
