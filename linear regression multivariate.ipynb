```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load data from Excel file
file_path = '/content/svm (1).xlsx'  # Replace with the actual file path
data = pd.read_excel(file_path)
X = data.iloc[:, :-1].values  # Use all columns except the last one as features
y = data.iloc[:, -1].values  # The last column is the target variable

# Step 2: Initialize model parameters
num_iterations = 1000
learning_rate = 0.01
m = len(y)
n = X.shape[1]  # Number of features
w = np.zeros(n)  # Initialize weights as an array of zeros

# Step 3: Implement gradient descent
for iteration in range(num_iterations):
    # Calculate predictions (hypothesis)
    y_pred = np.dot(X, w)

    # Calculate the cost (J(w))
    cost = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)

    # Calculate the gradient for all weights
    gradient = (1 / m) * np.dot(X.T, (y_pred - y))

    # Update weights using the gradient
    w = w - learning_rate * gradient

    # Print cost for every 100 iterations
    if iteration % 100 == 0:
        print(f'Iteration {iteration}: J(w) = {cost}')

# Step 4: Print the final weights and cost
print(f'Final weights: {w}')
print(f'Final cost: J(w) = {cost}')

# Step 5: There's no direct plot for multivariate regression, but you can plot the target variable against one feature
# Select one feature to create a scatter plot
selected_feature_index = 0  # Replace with the index of the feature you want to plot against the target
selected_feature = X[:, selected_feature_index]

plt.figure(figsize=(10, 6))
plt.scatter(selected_feature, y, label='Data Points', color='blue')
plt.plot(selected_feature, np.dot(X, w), label='Regression Line', color='red')
plt.xlabel(f'Feature {selected_feature_index}')
plt.ylabel('y')
plt.title('Multivariate Linear Regression with Gradient Descent')
plt.legend()
plt.grid(True)

plt.show()
```
