Conditional statements


```python
a=input("Enter the value of a")
b=input("Enter the value of b")
print(a)
print(b)
```

```python
if a>b:
  print("the value of a is greater than b")
if a==b:
  print("the value of a is equal to b")
else:
  print("The value of a is less than b")
```

```python
ball1_color="green"
ball2_color="red"
ball3_color="white"
ball4_color="pink"
pick_ball=input("please pick a ball")
if pick_ball==ball1_color:
  print("Green ball is picked")
elif pick_ball==ball2_color:
  print("Red ball is picked")
elif pick_ball==ball3_color:
  print("White ball is picked")
else:
  print("Pink ball is picked")
```
