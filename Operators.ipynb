Bitwise

```python
print(8&12)
print(0xF&0xE)
print(8|12)
print(0xF|0xE)
print(8^12)
print(0xF^0xE)
```

Assignment operators

```python
a=65
print(a)
a+=4
print(a)
a-=10
print(a)
a*=10
print(a)
a%=10
print(a)
```

Identity and Membership operators

```python
x=["bengaluru","goa"]
y=["bengaluru","goa"]
z=x
print(z is x)
print(y is not x)
print("hubli"in x)
```
